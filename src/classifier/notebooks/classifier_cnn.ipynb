# Earthquake Classifier — 2D CNN on Spectrograms
Binary classification: earthquake (1) vs noise (0).  
All logic lives in `src/classifier/` — this notebook trains, evaluates, and analyses errors.

```python
import sys
from pathlib import Path

sys.path.insert(0, str(Path('../../..').resolve()))

import matplotlib.pyplot as plt
from src.classifier.train import train
from src.classifier.evaluate import compute_metrics, error_analysis

DATA = str(Path('../../../data').resolve())
CKPT = str(Path('../../../checkpoints/classifier_best.pt').resolve())
```

## Train
Skips retraining if you already have a checkpoint — just comment out the `train()` call.

```python
history = train(DATA, CKPT)
```

## Training curves

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

axes[0].plot(history['train_loss'], label='Train')
axes[0].plot(history['val_loss'],   label='Val')
axes[0].set_xlabel('Epoch')
axes[0].set_ylabel('BCE Loss')
axes[0].set_title('Loss')
axes[0].legend()

axes[1].plot(history['val_f1'])
axes[1].set_xlabel('Epoch')
axes[1].set_ylabel('F1')
axes[1].set_title('Validation F1')

plt.tight_layout()
plt.show()
```
